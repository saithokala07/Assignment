### The following Code gives the factorial of a number using recursion

```python
def factorial(n):
    if n == 1:
        return 1
    elif n >1:
        return n*factorial(n-1)
    else:
        return "Enter a valid number"

n = int(input("Enter a Number:"))
print(factorial(n))

```
